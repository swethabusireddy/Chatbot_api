### A finite set is always countable.
>
>---
>*Example:*
>
>> ### Let $A= \{10, 11, 12, 13, ... \}$
>
> #### Set A has infinitely many elements.
> What is the Cardinality of this set?
>

>
> #### *Definition:*
>>  A set is countable if its elements can be placed in one-to-one correspondence with the natural numbers.
>>
>> Set of natural numbers $\mathbb{N}$ are:
>>> ### $\mathbb{N} = \{1, 2, 3, ...\}$
> ---

### The cardinality of countable sets is the same as the cardinality of natural numbers.
> ---
> *Example:*
>>### One-to-one correspondence meanes that we can pair all elements of one set with the elements of another set.
>>> **Let $O = \{1, 3, 5, 7, ... \}$ be the set of positive odd integers.**
>>>
>>>**Let $E = \{2, 4, 6, 8, ... \}$ be the set of positive even integers.** 
>>
>> ### We can pair elements of set O with set E as:
>>  Set | Elements
-- | --
>> Set $O$ \| | 1 3 5 7  `...`
>> | $\downarrow$ $\downarrow$ $\downarrow$ $\downarrow$
>> Set $E$ \| | 2 4 6 8 `...`
>>>
>>> ----
>>> *The first element of Set O (1) is paired with the first element of Set E (2).*
>>>
>>> *The second element of Set O (3) is paired with the second element of Set E (4).*
>>>
>>> *The third element of Set O (5) is paired with the third element of Set E (6).*
>>>
>>> *The fourth element of Set O (7) is paired with the fourth element of Set E (8).*
>>>
>>> *And so forth...*
>>>
>>> ---
>> 
>> **NOTE:** Each element of Set O is paired with only one element of Set E, and vice versa.
> ---


### Bijective Functions and Countable Sets
> ---
> #### *Definition:*
>> ### Let $\mathbb{N} = \{1, 2, 3, ... \}$
>> Define a function **f(n) = n + 9**
>>
>> Then **f(n)** is an element of the set:
>>> ### $A = \{10, 11, 12, ... \} $
>>
>> Therefore, **$f(n) = n + 9$** is a bijective function between **$\mathbb{N} = \{1, 2, 3, ... \}$** and **$A = \{10, 11, 12, ... \}$**.
>>
>> ---
>> For every element in Set B, there is only one element in Set N in which the bijective function can be applied to get the element in Set B.
>>
>> ---
>> If there exists a bijective function between two sets, then the elements of the sets can  be placed in a one-to-one correspondence.
>>
>> ---
>> A set is countable if its elements can be placed in one-to-one correspondence with the set of Natural numbers using a bijective function. 
>> Therefore, the following set is countable:
>>> ### $A = \{10, 11, 12, ...\}$
> ---
> *Example:*
>> ### Show that the set of integers $\mathbb{Z}$ is countable.
>>> #### $\mathbb{Z} = \{ ..., -3, -2, -1, 0, 1, 2, 3, ... \}$

>>>>

>>>>> and so on.
>>>>>
>>>>> ---
>>>>>
>>>>> Therefore, $\mathbb{Q}$ is a countable set.

### Uncountable Sets
> ---
> ### The set of real numbers $\mathbb{R}$ is uncountable.
>> A real number can be writen as an infinite decimal string.
>>
>> ---
>> We can show that the set $\mathbb{R}$ is uncountable by using **Cantor's Diagonal Method.**
>>> Assume that we have a "complete" list of real numbers written in the following decimals form:
>>>
>>>>
>>>> `0.39684187...`
>>>>
>>>> `0.84536219...`
>>>>
>>>> `0.69374693...`
>>>>
>>>> `0.71493478...`
>>>>
>>>> `0.12137876...`
>>>>
>>> ---
>>> The list above is infinite, yet complete in the sense that every real number will eventually appear in the list at the same point.
>>>
>>> ---
>>> Next, we highlight the first digit of the first number,
>>>
>>> the second digit of the second number,
>>>
>>> the third digit of the third number, and so on.
>>>
>>>>
>>>> `0.[3]9684187...`
>>>>
>>>> `0.8[4]536219...`
>>>>
>>>> `0.69[3]74693...`
>>>>
>>>> `0.714[9]3478...`
>>>>
>>>> `0.1213[7]876...`
>>>>
>>>>  $\vdots$
>>>>
>>> ---
>>> Now, we construct a new number by adding 1 to the highlighted digit.  If the digit is 9, adding 1 to it will give digit 0.
>>>> 
>>>> `0.[3]9684187...` >>>  `4`
>>>>
>>>> `0.8[4]536219...` >>> `5`
>>>>
>>>> `0.69[3]74693...` >>> `4`
>>>>
>>>> `0.714[9]3478...` >>> `0`
>>>>
>>>> `0.1213[7]876...` >>> `8`
>>>>
>>>>  $\vdots$
>>>>
>>> ---
>>> The new number which we created (i.e., 0.45408...) is not going to be anywhere in the original "complete" list.
>>>
>>> This is because the new numbers differs in the first difit from the first number in the list, differs in the second sifit from the second number in the list, and so on...
>>>
>>> ----
>>> In conclusion, the elements of $\mathbb{R}$ cannot be placed in one-to-one correspondence with the elements in $\mathbb{N}$.
>>>
>>> Therefore, $\mathbb{R}$ is uncountable.
> ---

### Summary
> ---
>#### * Finite Sets are always countable.  Infinite sets may be countable or uncountable.
